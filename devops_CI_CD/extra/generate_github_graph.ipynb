```python
import matplotlib.pyplot as plt
import networkx as nx

# Crear el grafo
G = nx.Graph()

# Añadir nodos principales
G.add_node("devops_CI_CD", type="center")

# Añadir nodos secundarios (ramas)
branches = ["Ansible", "Terraform", "Jenkins", "Docker", "Scripts"]
for branch in branches:
    G.add_node(branch, type="branch")
    G.add_edge("devops_CI_CD", branch)

# Añadir hojas por rama
leaves = {
    "Ansible": ["Playbooks", "Inventories", "Configs"],
    "Terraform": ["Modules", "AWS", "GCP"],
    "Jenkins": ["Pipelines", "Jobs", "Linting"],
    "Docker": ["AWS Images", "GCP Images", "Compose"],
    "Scripts": ["Ephemeral", "EC2 Mgmt", "GCP Auth"]
}

for parent, children in leaves.items():
    for child in children:
        G.add_node(child, type="leaf")
        G.add_edge(parent, child)

# Usar un layout
pos = nx.spring_layout(G, seed=42)

# Dibujar el grafo
plt.figure(figsize=(12, 12))
nx.draw(
    G, pos,
    with_labels=True,
    node_size=5000,
    node_color="skyblue",
    font_size=12,
    font_weight="bold",
    edge_color="black",
    width=2
)
plt.title("DevOps CI/CD Flow", fontsize=14)

# Guardar imagen para subir
output_path = "./devops_CI_CD_flow.png"
plt.savefig(output_path, format="PNG")
plt.show()

output_path
```
